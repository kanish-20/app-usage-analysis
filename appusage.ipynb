```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('app_usage.csv')
```

```python
print(df.head())
print(df.info())
```

```python
top_apps = df['app_name'].value_counts().head(10)
top_apps.plot(kind='bar')
plt.title('Top 10 Most Used Apps')
plt.show()
```

```python
top_apps = df['app_name'].value_counts().head(10)
top_apps.plot(kind='pie', autopct='%1.1f%%', startangle=150)
plt.title('Top 10 Most Used Apps')
plt.ylabel('')
plt.show()
```

```python
top_apps = df['app_name'].value_counts().head(10)
top_apps.plot(kind='line', marker='o')
plt.title('Top 10 Most Used Apps')
plt.ylabel('Usage Count')
plt.grid(True)
plt.show()
```
